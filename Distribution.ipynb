```python
a=[]
for i in range(1000):
    a.append(np.random.randint(1,100))

b=[]
c=[]
for i in range(100):
    b.append(a.count(i)/1000)
    c.append(i+1)
```

```python
print(plt.scatter(c,b))
print(plt.plot(b))
```

```python
def function(x):#here x is random variable and this is a normal distribution function
    mean=15.568090
    std=2.757689
    a=0.399 #value of 1/underroot(2*pie)
    term1=-((x-mean)/std)**2
    return a*(2.73**term1)/std
```

```python
b=[]
c=[]
for i in range(7,26):
   b.append(i)
   c.append(function(i))
```

```python
plt.scatter(b,c)
```
